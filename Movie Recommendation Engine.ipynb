```python
import pandas as pd
# Load dataset to a data frame
data = pd.read_csv('movies.csv')
data.head()
```

```python
print(data.shape)
```

```python
# Create a new data frame with relevant columns only
df = data[['budget',	'genres',	'homepage' , 'id', 'keywords',	'original_language',	'original_title',	'overview',	'popularity',	'production_companies',  'production_countries',  'release_date', 'revenue',	'runtime',	'spoken_languages',	'status',	'tagline',	'title',	'vote_average',	'vote_count',	'cast', 	'crew',	'director']].copy()
df.head()
```

```python
genres_all_movies = [
    df.loc[i]['genres'].split(' ')
    for i in df.index
    if pd.notna(df.loc[i]['genres'])
]

genres = sorted(list(set(
    item for sublist in genres_all_movies for item in sublist
)))

genres
```

```python
len(genres)
```

```python
# Initialize lists for movie data as well as titles
full_data = []
movie_titles = []
```

```python
print(df)
```

```python
df.to_csv('test.csv')
```

```python
for i in df.index:
    title = df.loc[i]['original_title']
    homepage = df.loc[i]['homepage']
    genres_value = df.loc[i]['genres']

    title = title.strip() if pd.notna(title) else ""
    homepage = homepage.strip() if pd.notna(homepage) else ""
    genres_value = genres_value if pd.notna(genres_value) else ""

    movie_titles.append((title, i, homepage))

    movie_data = [
        1 if genre in genres_value.split('|') else 0
        for genre in genres
    ]

    movie_data.append(df.loc[i]['vote_average'])

    full_data.append(movie_data)
```

```python
import os
import json

print(movie_titles)

# Create folder if it does not exist
os.makedirs('./Data', exist_ok=True)

data_dump = './Data/movie_data.json'
titles_dump = './Data/movie_titles.json'

# Save movie feature data
with open(data_dump, 'w', encoding='utf-8') as f:
    json.dump(full_data, f, ensure_ascii=False)

# Save movie titles data
with open(titles_dump, 'w', encoding='utf-8') as f:
    json.dump(movie_titles, f, ensure_ascii=False)

print("Files saved successfully ✅")
```

```python
df['genres'] = df['genres'].fillna('')
df['genres'] = df['genres'].str.split(' ')
```

```python
user_input = input("Enter genres separated by comma: ")

user_genres = [g.strip() for g in user_input.split(',')]
```

```python
def calculate_score(movie_genres):
    return len(set(movie_genres) & set(user_genres))

df['match_score'] = df['genres'].apply(calculate_score)
```

```python
recommended = df[df['match_score'] > 0] \
                .sort_values(by=['match_score', 'vote_average'], 
                             ascending=False)

print(recommended[['title', 'genres', 'vote_average',  'homepage']].head(10))
```
